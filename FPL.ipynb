Shape attribute of a dataframe returns the dimensionality, nothing but the number of rows and the number of columns of the dataframe. Therefore, there have been a total of 511 players across 20 teams that played in the English Premier League 2019-2020 season. 

```python
elements_df.describe()
```

The describe method returns statistical quantities for all the numerical columns present in the dataframe. For example, lets look at the now_cost column. A player at an average costs 5.2 million and there exists a standard deviation of 1.2 million in the cost distribution, which signifies that the distribution is pretty narrow. In simpler words, most of the players in FPL have their values very close to the average cost 5.2 million. Looking at the 75% quantile, we can verify this fact. 75% of the players cost less than or equal to 5.5 million. In contrast, observe the total_points column. Having a high standard deviation of 47.69, the values are more loosely spread in the distribution.

Beware of using the describe method without any caution. This method would also include all those attributes in your dataframe which might only associate with certain rows. For example, the saves column would only associate with the goalkeepers in our dataset, but the statistical quantities given here would be calculated across all the players in the dataset: defenders, midfielders and forwards too. Well, this would not have been a problem if everyone had Kyle Walker abilities, would it?

Lets try to isolate the goalkeepers and then use the describe method

```python
elements_df[elements_df.position == 'GKP'].describe()
```

See the difference? saves now have a mean value of 58 per goalkeeper. Now obviously, that is across the whole season.

Now, if your aim is to become a Data Engineer and run all kinds of fancy Machine Learning models on your dataset, 'Null' or 'NA' is probably the scariest value for you. In order to get a holistic view of the null values present in your dataset, the isna() couples with any() methods does the job:

```python
elements_df.isna().any()
```

Aha! We found the one culprit, the news_added column. The rest are all nice, having no missing data. Basically, isna() is a method which returns True/False value for each and every cell in your dataframe. Appending any() would then return a True value per column, if any of the rows in a particular column have atleast one null value. 

Lets see how many rows in the news_added column have null values:

```python
elements_df.isna().sum()
```

83 rows out of the total 511 have null values in the news_added column.
